```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
import xgboost as xgb

# 1. 데이터 로드 및 준비
# 파일 경로를 본인의 환경에 맞게 수정하세요.
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/DED_Dataset_preprocessed.csv')
target_col = 'Dry Eye Disease'

X = df.drop(columns=[target_col])
y = df[target_col]

# 2. 피처 중요도 순위 산출 (RFE 사용)
print("피처 순위 계산 중 (RFE)...")
rfe_estimator = RandomForestClassifier(n_jobs=-1, random_state=42)
rfe = RFE(estimator=rfe_estimator, n_features_to_select=1, step=1)
rfe.fit(X, y)

# 순위별 피처 정리
ranking_df = pd.DataFrame({'Feature': X.columns, 'Rank': rfe.ranking_})
sorted_features = ranking_df.sort_values('Rank')['Feature'].tolist()

print(f"피처 우선순위: {sorted_features}")

# 3. 모델 및 하이퍼파라미터 그리드 정의
models_params = {
    'DecisionTree': (DecisionTreeClassifier(random_state=42), {
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5]
    }),
    'RandomForest': (RandomForestClassifier(random_state=42, n_jobs=-1), {
        'n_estimators': [50, 100],
        'max_depth': [5, 10, None],
        'min_samples_split': [2, 5]
    }),
    'XGBoost': (xgb.XGBClassifier(eval_metric='logloss', use_label_encoder=False, random_state=42, n_jobs=-1), {
        'n_estimators': [50, 100],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5]
    }),
    'GradientBoosting': (GradientBoostingClassifier(random_state=42), {
        'n_estimators': [50, 100],
        'learning_rate': [0.05, 0.1],
        'max_depth': [3, 5]
    })
}

# 4. Feature 수 1개부터 N개까지 반복하며 최적 모델 탐색
results = []

# scoring 파라미터에 두 가지 지표를 모두 설정
scoring_metrics = {'accuracy': 'accuracy', 'f1': 'f1'}

# 전체 피처 다 돌리기 부담스러우면 range 범위를 조절하세요 (예: range(1, 11))
for k in range(1, len(sorted_features) + 1):
    current_features = sorted_features[:k]
    X_subset = X[current_features]

    print(f"Progress: Top {k} Features...")

    for model_name, (model, params) in models_params.items():
        # Grid Search 수행 (refit=False로 설정하여 재학습 시간 절약, 결과만 추출)
        grid = GridSearchCV(
            estimator=model,
            param_grid=params,
            cv=3,
            scoring=scoring_metrics, # accuracy와 f1 모두 계산
            refit=False,             # 최적 모델로 다시 학습하지 않음 (점수만 필요하므로)
            n_jobs=-1,
            verbose=0
        )
        grid.fit(X_subset, y)

        # 각 지표별 최고 점수 추출
        # cv_results_에는 각 파라미터 조합별 점수가 들어있음
        mean_test_acc = grid.cv_results_['mean_test_accuracy']
        mean_test_f1 = grid.cv_results_['mean_test_f1']

        best_acc = mean_test_acc.max()
        best_f1 = mean_test_f1.max()

        # 결과 저장
        results.append({
            'Num_Features': k,
            'Model': model_name,
            'Best_Accuracy': best_acc,
            'Best_F1': best_f1
        })

# 결과 데이터프레임 생성
results_df = pd.DataFrame(results)

# 5. 결과 시각화 (그래프 2개 그리기: Accuracy, F1)
fig, axes = plt.subplots(2, 1, figsize=(14, 14))

# 5-1. Accuracy 그래프
sns.lineplot(
    data=results_df,
    x='Num_Features',
    y='Best_Accuracy',
    hue='Model',
    marker='o',
    linewidth=2.5,
    ax=axes[0]
)
axes[0].set_title('Best Accuracy by Feature Count (Hyperparameter Tuned)', fontsize=16)
axes[0].set_ylabel('Best CV Accuracy', fontsize=12)
axes[0].legend(bbox_to_anchor=(1.01, 1), loc='upper left')
axes[0].grid(True, linestyle='--', linewidth=0.5)

# Accuracy 최고점 표시
best_acc_row = results_df.loc[results_df['Best_Accuracy'].idxmax()]
axes[0].scatter(best_acc_row['Num_Features'], best_acc_row['Best_Accuracy'], color='red', s=100, zorder=5)
axes[0].text(best_acc_row['Num_Features'], best_acc_row['Best_Accuracy'] + 0.005,
             f"Max Acc: {best_acc_row['Best_Accuracy']:.4f}\n({best_acc_row['Model']})",
             ha='center', color='red', fontweight='bold')

# 5-2. F1 Score 그래프
sns.lineplot(
    data=results_df,
    x='Num_Features',
    y='Best_F1',
    hue='Model',
    marker='o',
    linewidth=2.5,
    ax=axes[1]
)
axes[1].set_title('Best F1 Score by Feature Count (Hyperparameter Tuned)', fontsize=16)
axes[1].set_xlabel('Number of Features', fontsize=12)
axes[1].set_ylabel('Best CV F1 Score', fontsize=12)
axes[1].legend(bbox_to_anchor=(1.01, 1), loc='upper left')
axes[1].grid(True, linestyle='--', linewidth=0.5)

# F1 최고점 표시
best_f1_row = results_df.loc[results_df['Best_F1'].idxmax()]
axes[1].scatter(best_f1_row['Num_Features'], best_f1_row['Best_F1'], color='red', s=100, zorder=5)
axes[1].text(best_f1_row['Num_Features'], best_f1_row['Best_F1'] + 0.005,
             f"Max F1: {best_f1_row['Best_F1']:.4f}\n({best_f1_row['Model']})",
             ha='center', color='red', fontweight='bold')

plt.tight_layout()

# 그래프 및 결과 저장
plt.savefig('best_model_search_acc_and_f1.png')
results_df.to_csv('hyperparameter_tuning_results_acc_f1.csv', index=False)

print("완료! 그래프와 CSV 파일이 저장되었습니다.")
print(f"최고 Accuracy: {best_acc_row['Best_Accuracy']:.5f} ({best_acc_row['Model']}, Feats: {best_acc_row['Num_Features']})")
print(f"최고 F1 Score: {best_f1_row['Best_F1']:.5f} ({best_f1_row['Model']}, Feats: {best_f1_row['Num_Features']})")

plt.show()
```
